```python
import pandas as pd
import pymysql
import numpy as np
from sqlalchemy import create_engine
```

### 从数据库读取包裹信息

```python
# u_name="root"
pwd="root"
db="my_db"
host="localhost"
read_table="i_od_parcel_2025v31_mix_test"

divisor = 100 # 除数
# 写入的数据库
sql_table_name=r'i_od_parcel_2025v31_mix_pct1_test'

engine = create_engine('mysql+pymysql://{u_name}:{pwd}@{host}/{db}?charset=utf8'.format(
    u_name=u_name,
    pwd=pwd,
    db=db,
    host=host)
                      )
od_parcel = pd.read_sql_table(read_table,con=engine)
```

```python
# 所有irregular类型的parcel_id
irregular_small_id_list=list(set(list(od_parcel.loc[od_parcel.parcel_type.isin(["irregular"]),["parcel_id"]].parcel_id)))
# 所有isb类型的parcel_id
isb_small_id_list=list(set(list(od_parcel.loc[od_parcel.parcel_type.isin(["isb"]),["parcel_id"]].parcel_id))) 
# 所有nc类型的parcel_id
nc_small_id_list=list(set(list(od_parcel.loc[od_parcel.parcel_type.isin(["nc"]),["parcel_id"]].parcel_id))) 
# 所有parcel类型的parcel_id
parcle_small_id_list=list(set(list(od_parcel.loc[od_parcel.parcel_type.isin(["parcel"]),["parcel_id"]].parcel_id))) 
# 所有small类型的parcel_id
small_small_id_list=list(set(list(od_parcel.loc[od_parcel.parcel_type.isin(["small"]),["parcel_id"]].parcel_id))) 
```

```python
od_parcel_new=od_parcel.loc[:,['small_id','parcel_id','parcel_type']]
```

```python
len(od_parcel_new.loc[od_parcel_new.parcel_type.isin(['small']),'small_id'])
```

```python
a=od_parcel_new.groupby('parcel_type')
```

```python
type_dict=a.parcel_id.nunique().to_dict()
type_dict
```

```python
# od_parcel_new.parcel_type.nunique()
```

```python
irregular_list=[]
isb_list=[]
nc_list=[]
parcel_list=[]
small_list=[]
# 总的parcel_id
name_dic={'irregular': irregular_small_id_list,
          'isb': isb_small_id_list, 
          'nc': nc_small_id_list, 
          'parcel': parcle_small_id_list,
          'small': small_small_id_list}
# 存储取出的parcel_id
name_list={'irregular': irregular_list, 'isb': isb_list, 'nc': nc_list, 'parcel': parcel_list, 'small': small_list}
```

```python
import random
for k,val in type_dict.items():

    while True:
        if len(set(name_list[k]))==round(val/divisor):
            break
        parcel_id=random.choice(name_dic[k])
        name_list[k].append(parcel_id)
```

```python
parcel_li=irregular_list+isb_list+nc_list+parcel_list+small_list
```

```python
len(parcel_li)
```

```python
parcel_100=od_parcel.loc[od_parcel.parcel_id.isin(parcel_li),:]
```

```python
parcel_100
```

```python
parcel_100.to_sql(sql_table_name,engine,if_exists='append',chunksize=10000,index=False)
```
